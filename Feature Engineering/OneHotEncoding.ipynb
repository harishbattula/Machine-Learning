# One Hot Encoding using pandas and numpy

### It is an essential process of converting the categorical data variables to be provided to machine and deep learning algorithms which in turn improve predictions as well as classification accuracy of a model

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("benz.csv",usecols=["X2","X3","X5","X1"])
data.head()
```

```python
for column in data.columns:
    print(column,len(data[column].unique()))
```

```python
pd.get_dummies(data).shape
```

We can see 4 columns are encoded to 107 columns.
What if we have many unique values in each column and no of columns are n.
Then no of columns going to be very large whcih leads to increase in dimensionality of dataset and accuracy of model decreases.
To avoid this, we can seclect most frequent top m values in each column.

```python
# Getting all unique values in column and total no of apperances in each column
for column in data.columns:
    print(data[column].value_counts().sort_values(ascending=False))
```

```python
# Select top m values that are most freqent in each column
# Getting top m values for each column in dataset
Top10Values=[]
m=10
for column in data.columns:
    Top10Values.append(data[column].value_counts().sort_values(ascending=False).head(m).index)
    
# Top10Values=pd.DataFrame(Top10Values)
Top10Values
```

```python
cols=data.columns
for i in range(data.shape[1]):
    # without using dummy variable trap
    for label in Top10Values[i]:
        if(label!=None):
            data[cols[i]+"_"+label]=np.where(data[cols[i]]==label,1,0)
```

```python
data.drop(columns=cols,inplace=True)
data.shape
```
