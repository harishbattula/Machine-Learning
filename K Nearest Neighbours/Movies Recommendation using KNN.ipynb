```python
import pandas as pd
```

```python
# Loading data from csv file into a datframe
# It's sample data created from imdb..
df=pd.read_csv("Moviesdata.csv")
df
```

```python
# Converting string data type to numerical data so that it suits for mathematical computations
df=df.replace({
    "Hindi":0,
    "Telugu":1,
    "Kannada":2,
    "Tamil":3,
    "Malayalam":4
})
df
```

```python
# Getting valued from dataframe into array
data=df.values
```

```python
# Distance function that calculates distance from sample point to our interest point
def Distance(sampledata,interests):
    SumOfDistances=0
    for i in range(len(sampledata)):
        SumOfDistances+=abs(sampledata[i]-interests[i])
    return SumOfDistances
```

```python
# Creating KNN model that recommends movies to user
# But 
def RecommendMovies(data,Interests,NoOfMovies):
    
    RecommendationLabels=[]
    
    for index,sample in enumerate(data[:,1:-1]):
        
        RecommendationLabels.append((Distance(sample[:-1],Interests),index))
    
    # sorting Recommended lables
    # And getting best k recommendations out of it
    return [data[index][0] for distance,index in sorted(RecommendationLabels)[:NoOfMovies]]
```

### At scale, this would look like recommending products on Amazon, articles on Medium, movies on Netflix, or videos on YouTube. Although, we can be certain they all use more efficient means of making recommendations due to the enormous volume of data they process.

```python
Recommendations=RecommendMovies(data,[8.3,0,0,0,0,1,0,0,0],3)
Recommendations
```
